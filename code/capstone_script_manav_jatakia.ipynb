```python
#Standardizing column names
data.columns = data.columns.str.lower().str.replace(' ', '_')
```

```python
#Changing datatypes
# Clean and convert specified columns to int64
columns_to_convert = ['total_cost_(sub_projects)', 'funding_ic(s)', 'direct_cost_ic', 'indirect_cost_ic']

for column in columns_to_convert:
    data[column] = pd.to_numeric(data[column], errors='coerce').fillna(0).astype(int)
```

```python
import matplotlib.pyplot as plt

# Group data by fiscal year and calculate total cost
total_cost_by_year = data.groupby('fiscal_year')['total_cost'].sum()

# Create a bar chart
plt.figure(figsize=(8, 6))
total_cost_by_year.plot(kind='bar', color='skyblue')
plt.xlabel('Fiscal Year')
plt.ylabel('Total Cost (in billions)')
plt.title('Total Cost by Fiscal Year')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Group data by fiscal year
grouped_data = data.groupby('fiscal_year')

# Calculate number of projects and total cost for each fiscal year
summary_table = grouped_data.agg(
    number_of_projects=('project_title', 'count'),
    total_cost=('total_cost', 'sum')
).reset_index()

# Format total_cost column with commas and $ sign
summary_table['total_cost'] = summary_table['total_cost'].apply(lambda x: '${:,.2f}'.format(x))

print(summary_table)
```

```python
import pandas as pd

# Assuming 'data' is your DataFrame containing the dataset
# Assuming 'fiscal_year' is the column containing fiscal year information
# Assuming 'total_cost' is the column containing total cost information
# Assuming 'total_cost_ic' is the column containing total cost ic information

# Group data by fiscal year
grouped_data = data.groupby('fiscal_year')

# Calculate number of projects, total_fund, and total_cost for each fiscal year
summary_table = grouped_data.agg(
    number_of_projects=('project_title', 'count'),
    total_fund=('total_cost', 'sum'),
    total_cost=('total_cost_ic', 'sum')
).reset_index()

# Format total_cost and total_fund columns with commas and $ sign
summary_table['total_fund'] = summary_table['total_fund'].apply(lambda x: '${:,.2f}'.format(x))
summary_table['total_cost'] = summary_table['total_cost'].apply(lambda x: '${:,.2f}'.format(x))

print(summary_table)
```

```python
import matplotlib.pyplot as plt

#Barplot Discussion 6

# Group by 'administering_ic' and sum the 'total_cost_ic' if there are duplicates
grouped_data = data.groupby('administering_ic').sum().reset_index()

# Sort the DataFrame by 'total_cost_ic' in descending order and get the top 5
top_5 = grouped_data.sort_values(by='total_cost_ic', ascending=False).head(5)

# Convert 'total_cost_ic' to millions
top_5['total_cost_ic'] = top_5['total_cost_ic'] / 1e6

# Plotting the barplot
plt.figure(figsize=(10, 6))
plt.bar(top_5['administering_ic'], top_5['total_cost_ic'], color='skyblue')
plt.xlabel('Administering IC')
plt.ylabel('Total Cost (in millions)')
plt.title('Top 5 Administering ICs by Total Cost')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of projects for each 'administering_ic'
project_counts = data.groupby('administering_ic').count().reset_index()

# Sort the DataFrame by 'project_title' count in descending order and get the top 5
top_5_projects = project_counts.sort_values(by='project_title', ascending=False).head(5)

# Plotting the barplot
plt.figure(figsize=(10, 6))
plt.bar(top_5_projects['administering_ic'], top_5_projects['project_title'], color='skyblue')
plt.xlabel('Administering IC')
plt.ylabel('Number of Projects')
plt.title('Top 5 Administering ICs by Project Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting numerical columns for correlation
numerical_cols = ['application_id', 'fiscal_year', 'total_cost', 'total_cost_ic', 
                  'total_cost_(sub_projects)', 'direct_cost_ic', 'indirect_cost_ic']

# Calculating the correlation matrix
correlation_matrix = data[numerical_cols].corr()

# Plotting the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Ensure 'Fiscal Year' and 'Total Cost' columns exist in the dataset
if 'fiscal_year' in data.columns and 'total_cost' in data.columns:
    # Aggregate the total cost by year
    data_aggregated = data.groupby('fiscal_year')['total_cost'].sum().reset_index()

    # Fit the Exponential Smoothing model
    model = ExponentialSmoothing(data_aggregated['total_cost'], trend='add', seasonal=None, seasonal_periods=None)
    fit = model.fit()

    # Forecast the Total Cost for 2024
    forecast_2024 = fit.forecast(steps=1)

    print("Forecasted Total Cost for 2024: $", forecast_2024.values[0])
```
